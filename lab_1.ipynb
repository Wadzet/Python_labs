```python
import time

def delay_print(s):
    for c in s:
        print(c, end='')
        time.sleep(0.05)
    print('')


def start_game():
    answers_array = []
    quest_index = 0

    print('Ватаємо у "Mafia: Demo"! Виберіть свого героя:'
    "\n1) Джонні"
    "\n2) Антоніо")
    
    answer = input("Виберіть героя: ")
    answers_array.append(answer)
    
    match answer:
        case "1":
            print("Ви обрали Мафіозі Джонні.")
            delay_print("Джонні: 'Ми вже на шляху до нашої нової операції.'")
            quest_index += 1
            
        case "2":
            print("Ви обрали Мафіозі Антоніо.")
            delay_print("Антоніо: ' Почнемо нашу нову операцію.'")
            quest_index += 1
            
        case _:
            print("Невірний вибір. Спробуйте ще раз.")
            start_game()
    
    # Головна гілка гри
    while quest_index < 5:
        match quest_index:
            case 1:
                print("Ви відправляєтесь до місця проведення операції у Чайна-Таун. Це ресторан сіміʼї Карпоне, ви ніколи раніше тут не були."
                      "\n Вам потрібно забрати товар\nЯкі будуть ваші дії: "
                    "\n1) Почати перестрілку без роздумів"
                    "\n2) Обережно напасти з засідки")

                answer = input("Ваш вибір: ")
                answers_array.append(answer)
                
                if answer == "1":
                    delay_print("Ваша несподівана атака виявилася успішною. Ви вбили всіх паразитів Карпоне.")
                elif answer == "2":
                    delay_print("Ви вирішили напасти з засідки, але вас помітили. По вас відкрили вогонь."
                                "\nАле вам вдалось все таки добити псів Карпоне і забрати товар")
                quest_index += 1
                
            case 2:
                delay_print("Але сталась неочікувана ситуація, на це місце приїхали сучі Гамбіно. Їх кількість сильно більша за вашу."
                      "\nЇх головний пропонує вам ʼвигіднуʼ сдєлку. Ви нам товар - а ми вас відпустим. "
                      "Ви можете:" 
                      "\n1) Ризикнути і почати ʼмясиловоʼ"
                      "\n2) Погодитись на їх умови")
                
                answer = input("Ваш вибір: ")
                answers_array.append(answer)
                
                if answer == "1":
                    delay_print("Частина вашої сімʼї поранена, багато пристрілено, але вам вдалось придушити цих щурів і забрати товар.")
                    end_game(answers_array)

                elif answer == "2":
                    delay_print("Вас обдурили і почали вогонь, практично всі ваші загинули, але вам ще з 3ма братки якось вдалось вийти з ресторану і сісти в тачку."
                                "\nАле за вами почалась погоня.")
                quest_index += 1
                
            case 3:
                print("Який маршрут ви оберете."
                "\n1) Їхати на базу"
                "\n2) Заїжати по дворикам, постаратись відірватись")
                
       
                answer = input("Ваш вибір: ")
                answers_array.append(answer)
                
           
                if answer == "1":
                    delay_print("Ви заїхали на база і ваша сімʼя розібралась з тарганами.")
                elif answer == "2":
                    delay_print("Ви отрапили в ще гірший стан. За вами тепер ще три машини, які почали по вас вогонь і ви розбились.")
                    end_game(answers_array)
                    return
                quest_index += 1
                
            
                
                
 
    end_game(answers_array)

def end_game(answers):
    print("Гра завершилася.")
 
    if "2" in answers:
        print("Гейм овер. Ви здохли :(")
    elif "2" not in answers and "1" in answers:
        print("Вітаємо! Ви виграли!!!")
    else:
        print("Гра закінчилася.")
    
    play_again = input("Хочете зіграти ще раз? (Так/Ні): ").lower()
    if play_again == "так":
        start_game()
    else:
        print("Дякуємо за гру!")

start_game()
```
